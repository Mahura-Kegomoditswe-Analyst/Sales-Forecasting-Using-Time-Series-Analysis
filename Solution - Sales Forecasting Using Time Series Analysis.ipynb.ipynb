# Unpredictable Sales Fluctuations

```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('/Users/mdshahriar/Downloads/Python Projects/Project 1/sales_forecasting_data.csv', parse_dates=["date"])

# Sort data by date
df = df.sort_values("date")

# Plot sales trends
plt.figure(figsize=(12, 5))
plt.plot(df["date"], df["sales"], label="Daily Sales", color='blue', alpha=0.7)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Daily Sales Over Time")
plt.legend()
plt.show()
```

# Lack of Seasonal Insights

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform time series decomposition
decomposition = seasonal_decompose(df.set_index("date")["sales"], model="additive", period=7)

# Plot components
decomposition.plot()
plt.show()
```

### Code Explanation:

- **Imports**:
  - `from statsmodels.tsa.seasonal import seasonal_decompose`: Imports the `seasonal_decompose` function from `statsmodels`, which is used for time series decomposition to extract trend, seasonal, and residual components.

- **Time Series Decomposition**:
  - `decomposition = seasonal_decompose(df.set_index("date")["sales"], model="additive", period=7)`: 
    - Converts the `date` column to the index of the DataFrame.
    - Applies the `seasonal_decompose` function to the `sales` data, using an **additive** model (assumes that the time series is the sum of trend, seasonal, and residual components).
    - The `period=7` argument assumes that there is weekly seasonality (7 days in a week), and the function decomposes the time series accordingly.

- **Plotting the Decomposition Components**:
  - `decomposition.plot()`: Plots the decomposed components (Trend, Seasonal, Residual).
  - `plt.show()`: Displays the plot with the decomposed components.
